# Exploratory Analysis

This notebook explores frequency-domain differences between real and synthetic images.

```python
import os
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

sys.path.append("..")

from src.forensic_features import _fft_power_spectrum

DATA_DIR = Path("../data")
REAL_DIR = DATA_DIR / "real"
SYN_DIR = DATA_DIR / "synthetic"

real_paths = list(REAL_DIR.glob("*.jpg")) + list(REAL_DIR.glob("*.png"))
syn_paths = list(SYN_DIR.glob("*.jpg")) + list(SYN_DIR.glob("*.png"))

if not real_paths or not syn_paths:
    raise ValueError("Please add images to data/real and data/synthetic")

sample_real = Image.open(real_paths[0]).convert("L")
sample_syn = Image.open(syn_paths[0]).convert("L")

real_fft = _fft_power_spectrum(np.array(sample_real))
syn_fft = _fft_power_spectrum(np.array(sample_syn))

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(np.log1p(real_fft), cmap="magma")
plt.title("Real FFT Power")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(np.log1p(syn_fft), cmap="magma")
plt.title("Synthetic FFT Power")
plt.axis("off")
plt.tight_layout()
plt.show()
```
