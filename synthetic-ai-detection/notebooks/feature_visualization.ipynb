# Feature Visualization

This notebook visualizes forensic features and Grad-CAM explanations.

```python
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sys.path.append("..")

from src.forensic_features import build_feature_matrix, collect_image_paths

paths, labels = collect_image_paths("../data")
features = build_feature_matrix(paths)

feature_df = pd.DataFrame(features.features, columns=features.names)
feature_df["label"] = labels

plt.figure(figsize=(8, 4))
sns.boxplot(data=feature_df, x="label", y="high_freq_ratio")
plt.title("High-Frequency Ratio by Class")
plt.show()
```

```python
from pathlib import Path
from src.faithfulness_analysis import run_gradcam

cnn_model = Path("../outputs/models/cnn_resnet50.pt")
example_image = paths[0]

if cnn_model.exists():
    out_path = Path("../outputs/plots/gradcam_notebook.png")
    run_gradcam(example_image, str(cnn_model), out_path)
    print(f"Saved Grad-CAM to {out_path}")
else:
    print("Train the CNN model first to generate Grad-CAM outputs.")
```
